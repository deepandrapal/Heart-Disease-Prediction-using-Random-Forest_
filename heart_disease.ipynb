# Heart Disease Prediction

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
dataSet = pd.read_csv('heart-disease.csv')
dataSet.head()
```

```python
X = dataSet.iloc[:, :-1].values
y = dataSet.iloc[:, -1].values
```

## Split data into Train and Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

## Training The model

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=100)

clf.get_params()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

## Model optimisation

```python
np.random.seed(100)
for i in range(10, 150, 10):
    clf = RandomForestClassifier(n_estimators=i, random_state=100).fit(X_train, y_train)
    print(clf.score(X_test, y_test), i)
```

```python
clf = RandomForestClassifier(n_estimators=30, random_state=100)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

## Performance Testing

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred = clf.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
